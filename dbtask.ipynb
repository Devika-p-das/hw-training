```python
import pandas as pd
from pymongo import MongoClient
```

```python
file_path = "/home/user/Downloads/DataHut_CH_Denner_FullDump_20260618.CSV"

# Read pipe-delimited CSV
df = pd.read_csv(file_path, sep="|")

print("Rows:", len(df))
print("Columns:", len(df.columns))
```

```python
df = df.where(pd.notnull(df), None)

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Database
db = client["practice_db"]

# Collection
collection = db["coop"]

# Import data
collection.insert_many(df.to_dict("records"))

print(f"{len(df)} records imported successfully into practice_db.coop!")
```
