```python
# Inventory dictionary with item prices

inventory = {
    "GoodDay Biscuit": 30.5,
    "milk": 45.0,
    "bread": 25.5,
    "eggs": 60.0,
    "chocolate": 50.0,
    "chocolate": 55.0
}
```

```python
# Cart list with selected items
cart = ["apple", "milk", "bread", "chocolate","chocolate"]
# Print data types
print("Type of inventory:", type(inventory))
print("Type of one price value:", type(inventory["milk"]))
print("Type of cart:", type(cart))
```

```python
#Total bill calculation
total_bill = 0
for item in cart:
    if item in inventory:
        total_bill += inventory[item]
    else:
        print(f"Item '{item}' not found in inventory. Skipping.")

print(f"Total Bill: {total_bill}")
```

```python
#Using if-else to check if selected items exist in the inventory and Warning about unavailable items
for item in cart:    
    if item in inventory: 
        print(f"{item} is available at price {inventory[item]}") 
    else: 
        print(f"Warning: {item} is not available in the inventory.")
```

```python
#Converting the cart list into a set to show only unique items.
unique_cart = set(cart)
print("Unique items in the cart:", unique_cart)
```

```python
#Creating a tuple of product categories and printing it and its type.
product_categories = ("fruits", "dairy", "bakery")
print("Product categories:", product_categories)
print("Type of product_categories:", type(product_categories))
```

```python
# Add item with None price
inventory["chips"] = None

# Print the type of the None price
print("Type of chips price:", type(inventory["chips"]))
```

```python
# Add a boolean variable
is_discount_applied = False

# Set it to True if total bill > 100
total_bill = sum(inventory[item] for item in cart if item in inventory)
if total_bill > 100:
    is_discount_applied = True

print("Is discount applied?", is_discount_applied)
```
