```python
# Creates a 2x3 grid of subplots displaying box plots for all six integer columns.

int_cols = [
    'item_mrp', 'item_selling_price', 'product_price', 
    'promo_offer_price', 'discount_in_percent', 'stock_count'
]

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))

axes = axes.flatten()

for i, col in enumerate(int_cols):
    sns.boxplot(data=df, y=col, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Distribution of {col}', fontsize=12)
    axes[i].set_ylabel('')  # Clears y-label since title has the column name


plt.tight_layout()
plt.show()
```

###  BIVARIATE ANALYSIS

```python
# Calculates and groups the sorted average discount percentage for each department, rounded to two decimals.

dept_discount = (
    df.groupby('department')['discount_in_percent']
      .mean()
      .sort_values(ascending=False)
)
dept_discount.round(2)
```

```python
# Plots sorted average department discounts as an orange bar chart with centered percentage value labels.

plt.figure(figsize=(12,5))

dept_discount.plot(
    kind='bar',
    color='orange'
)

plt.title('Average Discount Percentage by Department')
plt.ylabel('Average Discount (%)')
plt.xlabel('Department')

for i, v in enumerate(dept_discount):
    plt.text(i, v+0.5, f'{v:.1f}%', ha='center')

plt.xticks(rotation=45)
plt.show()
```

```python
# Creates a frequency table crossing department names with out-of-stock statuses to show item availability counts.

pd.crosstab(
    data['department'],
    data['out_of_stock_flag']
)
```

```python
# Plots a grouped count chart comparing in-stock versus out-of-stock item frequencies across departments.

plt.figure(figsize=(12,5))

sns.countplot(
    data=data,
    x='department',
    hue='out_of_stock_flag'
)

plt.xticks(rotation=45)
plt.title('Department vs Out of Stock')
plt.show()
```

```python
# Prints and plots the sorted out-of-stock percentage per category using a horizontal lollipop chart.

oos_by_category = (
    df.groupby('category')['out_of_stock_flag']
      .mean()
      .mul(100)
      .sort_values(ascending=False)
)

print(oos_by_category.round(2))
plt.figure(figsize=(12,5))

plt.hlines(
    y=oos_by_category.index,
    xmin=0,
    xmax=oos_by_category.values
)

plt.plot(
    oos_by_category.values,
    oos_by_category.index,
    'o'
)

plt.xlabel('Out of Stock (%)')
plt.title('Out of Stock Percentage by Category')

plt.show()
```

```python
# Prints the top 20 and plots the top 15 most expensive brands as a horizontal skyblue bar chart.

brand_price = (
    df.groupby('brand_name')['product_price']
      .mean()
      .sort_values(ascending=False)
)

print(brand_price.head(20))
plt.figure(figsize=(12,6))

brand_price.head(15).sort_values().plot(
    kind='barh',
    color='skyblue'
)

plt.title('Top Brands by Average Product Price')
plt.xlabel('Average Price')
plt.show()
```

```python
# Plots discount distributions across the top 10 most frequent brands using a vertical box plot grid.

top_brands = df['brand_name'].value_counts().head(10).index

plt.figure(figsize=(14,6))

sns.boxplot(
    data=df[df['brand_name'].isin(top_brands)],
    x='brand_name',
    y='discount_in_percent'
)

plt.xticks(rotation=45)

plt.title('Brand vs Discount Distribution')
plt.show()
```

```python
# Plots the top 10 store locations with the highest average product price as a skyblue vertical bar chart.

store_price = (
    df.groupby('store_name_location')['product_price']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(12,5))

store_price.plot(
    kind='bar',
    color='skyblue'
)

plt.title('Stores by Average Product Price')
plt.ylabel('Average Price')
plt.show()
```

```python
# Plots the top 15 pin codes with the highest average discount percentage as an orange vertical bar chart.

pin_discount = (
    df.groupby('pin_code')['discount_in_percent']
      .mean()
      .sort_values(ascending=False)
)

plt.figure(figsize=(12,5))

pin_discount.head(15).plot(
    kind='bar',
    color='orange'
)

plt.title(' Pin Codes by Average Discount')
plt.ylabel('Average Discount (%)')
plt.show()
```

### MULTIVARIATE ANALYSIS

```python
# Calculates row counts for every unique combination of department, category, and sub-category as a flat table.

data.groupby(['department', 'category', 'sub_category']).size().reset_index(name='count')
```

```python
#  break down the average discount by department AND brand
brand_discount = (
    df.groupby(['department', 'brand_name'])['discount_in_percent']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)
brand_discount.round(2)
```

```python
# Plots product price versus discount percentage as a scatter plot, color-coded by the top 5 most frequent categories.

top_categories = df['category'].value_counts().head(5).index

plt.figure(figsize=(12,6))

sns.scatterplot(
    data=df[df['category'].isin(top_categories)],
    x='product_price',
    y='discount_in_percent',
    hue='category'
)

plt.title('Price vs Discount by Category')
plt.show()
```

```python
# Stores ,category and outoff stock percentage
pivot = df.groupby(['store_name_location', 'category'])['out_of_stock_flag'].mean().unstack()*100

pivot = pivot.round(2).astype(str) + '%'

pivot
```

```python
# Plots a pin code scatter plot of average price versus average discount, labeling each data point with its pin code.

pin_summary = df.groupby('pin_code').agg({
    'product_price':'mean',
    'discount_in_percent':'mean'
}).reset_index()

plt.figure(figsize=(10,6))
sns.scatterplot(data=pin_summary, x='product_price', y='discount_in_percent')

for idx, row in pin_summary.iterrows():
    plt.text(row['product_price'] + 0.3, row['discount_in_percent'], str(row['pin_code']), fontsize=9)

plt.title('Pin Code Average Price vs Average Discount')
plt.show()

```

```python
# Computes a linear correlation matrix for five numerical columns and plots it as an annotated heatmap.

num_cols = [
    'item_mrp',
    'item_selling_price',
    'product_price',
    'discount_in_percent',
    'stock_count'
]

plt.figure(figsize=(8,6))

sns.heatmap(
    df[num_cols].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Matrix')
plt.show()
```

```python
# Plots a brand-level scatter plot comparing average product price against average discount percentage.

brand_summary = (
    df.groupby('brand_name')
      .agg({
          'product_price':'mean',
          'discount_in_percent':'mean'
      })
      .reset_index()
)

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=brand_summary,
    x='product_price',
    y='discount_in_percent'
)

plt.xlabel('Average Price')
plt.ylabel('Average Discount (%)')
plt.title('Brand Price vs Discount')
plt.show()
```
