```python
import pandas as pd
from pymongo import MongoClient
```

```python
file_path = "/home/user/Downloads/DataHut_AT_Dm_PriceExtractions_20260619.CSV (1)/DataHut_AT_Dm_PriceExtractions_20260619.CSV"

# Read pipe-delimited CSV
df = pd.read_csv(file_path, sep="|")

print("Rows:", len(df))
print("Columns:", len(df.columns))
```

```python
df = df.where(pd.notnull(df), None)

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Database
db = client["practice_db"]

# Collection
collection = db["Dm20260619"]

# Import data
collection.insert_many(df.to_dict("records"))

print(f"{len(df)} records imported successfully into practice_db.Dm20260619!")
```
